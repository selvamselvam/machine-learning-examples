```python
import time
import datetime

print(time.mktime(datetime.datetime.strptime("2021-01-14",
                                             "%Y-%m-%d").timetuple()))
```

```python
X = dffilterindia[[ 'total_cases',  'hospital_beds_per_thousand',
       'female_smokers', 'male_smokers']]
```

```python
y = dffilterindia['new_cases']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print("intercept:",lm.intercept_)
print("COEF:",lm.coef_)
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```

```python
predications = lm.predict(X_test)
```

```python
plt.scatter(y_test,predications)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predications))
print('MSE:', metrics.mean_squared_error(y_test, predications))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predications)))
```
